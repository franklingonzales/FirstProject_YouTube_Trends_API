```python
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
df = pd.read_csv('resources/avg_rain_state.csv')

df.head(5)
```

```python
df['Inches'].min()
```

```python
df['Inches'].max()
```

```python
plt.rcParams['figure.figsize'] = [15, 5]

plt.bar(df['State'], df['Inches'], color='red', alpha=0.3, align="center") 

plt.title('Average Rain per State',fontsize=20)

plt.xlabel('State')
plt.ylabel('Inches')
plt.xticks(rotation='vertical')
plt.ylim(9, 65)

plt.tight_layout()
plt.show()
```
